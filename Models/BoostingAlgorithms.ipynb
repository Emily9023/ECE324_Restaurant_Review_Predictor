```python
import json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn import metrics
```

```python
df = pd.read_csv("cleanedData_cuisinevec.csv", sep=',')
# df = df.rename(index={0: "stars", 1: "delivery", 2: "takeout", 3: "outdoor seating", 4: "closingTime", 5: "cost", 6: "cuisine", 7: "location"})
df
```

```python
# Perform train-test split
Y_col = 'stars'
X_cols = df.loc[:, df.columns != Y_col].columns
X_train, X_test, y_train, y_test = train_test_split(
    df[X_cols], df[Y_col],test_size=0.1, random_state=42)
y_train = list(y_train)
y_test = list(y_test)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
xgboost=XGBClassifier()
xgboost.fit(X_train,to_categorical(y_train))
xgboost.score(X_test, to_categorical(y_test))
```

```python
model = GradientBoostingClassifier(
    n_estimators=100, learning_rate=1, max_depth=1, random_state=42)

str_y_train = list(map(str, y_train))
str_y_test = list(map(str, y_test))

model.fit(X_train, (str_y_train))
print(model)
model.score(X_test, (str_y_test))
```

For false-negative calculations:

```python
pred = model.predict(X_test)
pred
```

```python
# for FNR

pred_binary = []
for e in pred:
  if float(e) >= 4:
    pred_binary.append(1)
  else:
    pred_binary.append(0)
  # print(pred_binary)

y_test_binary = []
for e in y_test:
  # print(e)
  if float(e) >= 4:
    y_test_binary.append(1)
  else:
    y_test_binary.append(0)
  # print(y_test_binary)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test_binary, pred_binary)
confusion_matrix
```
